```python
#Solar Panel Dirtying

import matplotlib.pyplot as plt
def SolarPanel(days):
    SimulationData = []
    SimulationDays = []
    efficiency = 0.3
    degradation =efficiency/100
    i = 0
    while i < days:
        SimulationData.append(efficiency)
        SimulationDays.append(i)
        efficiency = efficiency - degradation
        i = i + 1
    plt.plot(SimulationDays, SimulationData)
    plt.show()
    
SolarPanel(365)    
    
    

    
```

```python
#Combining Solar Panel and ClimateData

import os 
import csv
import matplotlib.pyplot as plt
import numpy as np

Date = []
Rain = []
CVal=[]
CleanData = []
NoCleanData = []
SimulationDays = []
Date1=[]
Radiation=[]
Radiation2=[]
MaxRadiation = []
PanelPower=[]
Eta=[]


efficiencyNoClean = 1
A_Pan = 5.4*3.25*.3048
A_Par = 1.257*10**(-9)
P_Cur = 10*A_Pan
P_Pan =[]
cx = 5.4*.3048
lamda = 1.9/1000
n = 2.83 
f = 10 
V=np.linspace(0,0.9999999,100001)
degradation = 0.0014063 #What I'll be subtracting from the efficiency






with open("ClimateData2019.csv") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    csv_list = list(csv_reader)
    line_count = 5
    data =[]
    for row in csv_list:
        if 0 < csv_list.index(row):
            Rain.append(row[3])
            Date.append(row[2])


i = 0




days = len(Date)
i = 0


    
with open("RadiationData.csv") as csv_file1:
    csv_reader1 = csv.reader(csv_file1, delimiter=',')
    csv_list = list(csv_reader1)
    line_count = 5
    data =[]
    for row in csv_list:
        if csv_list.index(row)> 2:
            Radiation.append(row[6])

Radiation = list(map(int, Radiation)) #Converting list of strings to ints

#Break up data into days
for DHI in Radiation:
    if i+48 <= len(Radiation):
        Radiation2.append(Radiation[i:i+48])
        
    i = i+48
#Assume efficiency of solar panel is 15%
#Pick out the maximum radiation for every day throughout the year and use that for the power input to panel
for maximum in Radiation2:
    MaxRadiation.append(max(maximum))
    PPower = max(maximum)*0.15
    P_Pan.append(PPower)

    
#Find the value for C for all values of P that were given
for P in P_Pan:     
    C= (A_Par/A_Pan)*(P_Cur/P)*(cx/lamda)*(n/f)
    CVal.append(C)
   
   
    
totalC = sum(CVal)
CAverage = totalC/len(CVal)   
#plt.plot(SimulationDays, SimulationData)
#plt.xlabel("Days")
#plt.ylabel("Efficiency")
#plt.show()
for Voltage in V: 
    eta1 = 0.5*(1+Voltage**2)
    eta = eta1 -(CAverage/(1-Voltage**2))
    Eta.append(eta)

maxEta = max(Eta)


maxEtaLocation = Eta.index(maxEta)
V_On = V[maxEtaLocation]
#print("This is the voltage that the electrodynamic screen should turn on: "+ str(V_On))
#print("This is the maximum efficiency of the solar panel with the shield: " + str(max(Eta)))
#print(len(Eta))
#plt.plot(V,Eta)


#Efficiency Loop
counter = 0
efficiencyClean = maxEta
VoltageClean = 1
while counter < days:
    
    NoCleanData.append(efficiencyNoClean)
    CleanData.append(efficiencyClean)
    if float(Rain[counter]) == 0:
        efficiencyNoClean = efficiencyNoClean - degradation
        efficiencyClean = efficiencyClean - degradation
        VoltageClean = VoltageClean - degradation
        if VoltageClean < V_On:
            efficiencyClean = maxEta
            VoltageClean = 1
            
            
    else:
        efficiencyNoClean = 1
        efficiencyClean = maxEta
        VoltageClean=1
    counter += 1
#Power Loop    
CleanPower=[]
NoCleanPower=[]
counter2 = 0    
for Power in P_Pan:
    ScreenPower = Power * CleanData[counter2]
    NoPower = Power * NoCleanData[counter2]
    
    CleanPower.append(ScreenPower)
    NoCleanPower.append(NoPower)
    counter2 += 1
    
    
    
#Power Difference    
difference = 0   
DiffEnergy = []
while difference < 365:
    CleanEnergy = (CleanPower[difference]*A_Pan*86400)/3600000
    DirtyEnergy = (NoCleanPower[difference]*A_Pan*86400)/3600000
    diff = CleanEnergy-DirtyEnergy
    DiffEnergy.append(diff)
    difference += 1

TotalEnergySaved = sum(DiffEnergy)


#Using electric rate of $0.20/kWh
Cost = TotalEnergySaved * .20
print('This is the total energy difference: ' + str(TotalEnergySaved) +"kWh")
print('This is the total cost: $' + str(Cost))

#Plots
plt.plot(Date,NoCleanData)
plt.title("No Cleaning Method")
plt.ylabel("Percent Efficiency")
plt.xlabel("Date")
plt.xticks([0,30,58,89,119,150,180,211,242,272,303,333],['1-1', '2-1', '3-1', '4-1', '5-1', '6-1',
                                                   '7-1', '8-1', '9-1', '10-1', '11-1', '12-1'])
plt.show()
          
plt.plot(Date,CleanData)
plt.title("Electrodynamic Screen")
plt.ylabel("Percent Efficiency")
plt.xlabel("Date")
plt.xticks([0,30,58,89,119,150,180,211,242,272,303,333],['1-1', '2-1', '3-1', '4-1', '5-1', '6-1',
                                                   '7-1', '8-1', '9-1', '10-1', '11-1', '12-1'])
plt.show()
          



plt.plot(Date,NoCleanPower, label='Bare Solar Panel')
plt.plot(Date,CleanPower,'--', label='Solar Panel with Electrodynamic Screen')
plt.title("Power Output of Solar Panel")
plt.ylabel("Power Output(W/m^2)")
plt.xlabel("Date")
plt.xticks([0,30,58,89,119,150,180,211,242,272,303,333],['1-1', '2-1', '3-1', '4-1', '5-1', '6-1',
                                                   '7-1', '8-1', '9-1', '10-1', '11-1', '12-1'])
plt.legend()
plt.legend()
plt.show()

plt.plot(Date,NoCleanPower, label='Bare Solar Panel')
plt.plot(Date,CleanPower,'--', label='Solar Panel with Electrodynamic Screen')
plt.fill_between(Date,NoCleanPower,CleanPower,color = 'red')
plt.title("Power Output of Solar Panel")
plt.ylabel("Power Output(W/m^2)")
plt.xlabel("Date")
plt.xticks([0,30,58,89,119,150,180,211,242,272,303,333],['1-1', '2-1', '3-1', '4-1', '5-1', '6-1',
                                                   '7-1', '8-1', '9-1', '10-1', '11-1', '12-1'])
plt.legend()
plt.show()

```
