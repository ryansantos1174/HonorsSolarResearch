```python
#Solar Panel Dirtying

import matplotlib.pyplot as plt
def SolarPanel(days):
    SimulationData = []
    SimulationDays = []
    efficiency = 0.3
    degradation =efficiency/100
    i = 0
    while i < days:
        SimulationData.append(efficiency)
        SimulationDays.append(i)
        efficiency = efficiency - degradation
        i = i + 1
    plt.plot(SimulationDays, SimulationData)
    plt.show()
    
SolarPanel(365)    
    
    

    
```
